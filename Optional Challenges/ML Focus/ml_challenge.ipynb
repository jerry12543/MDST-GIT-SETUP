```python
# you are free to use any package you deem fit
import numpy as np
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
```

## Dataset

```python
# DO NOT MODIFY
from sklearn.datasets import make_moons

X, Y = make_moons(random_state=42, n_samples=(50, 450), noise=0.25)
```

## Training

```python
# want to classify between 0s and 1s (moon 1 and moon 0) given the X values

#show the moons
plt.xlabel("col 0")
plt.ylabel("col 1")
plt.scatter(X[:,0][Y==0], X[:,1][Y==0])
plt.scatter(X[:,0][Y==1], X[:,1][Y==1])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = .4, random_state = 42)
```

```python
from sklearn import tree#LogisticRegression
model = tree.DecisionTreeClassifier() 
# 1st test = LogisticRegression, .955 and .93, missing low hanging 0s
# 2nd test = DecisionTreeClassifier, .935, .974
model.fit(X_train, y_train)
```

## Testing / Optimization

```python
y_pred = model.predict(X_test)
print(model.score(X_test, y_test))
print(np.mean(model.predict(X_test)-y_test))
print(np.mean(np.abs(model.predict(X_test)-y_test)))
```

```python
#plt.scatter(X_test[:,0][y_test==0], X_test[:,1][y_test==0])
#plt.scatter(X_test[:,0][y_test==1], X_test[:,1][y_test==1])
plt.scatter(X_test[:,0][y_pred==0], X_test[:,1][y_pred==0])
plt.scatter(X_test[:,0][y_pred==1], X_test[:,1][y_pred==1], color = "c")
plt.scatter(X_test[:,0][y_pred!=y_test], X_test[:,1][y_pred!=y_test], color = "r") # red dots indicate incorrect classification
```

```python
#Test model on 2nd set of moons

W, Z = make_moons(random_state=42, n_samples=(50, 450), noise=0.25)
z_pred = model.predict(W)
print(model.score(W,Z))
plt.scatter(W[:,0][Z==0], W[:,1][Z==0])
plt.scatter(W[:,0][Z==1], W[:,1][Z==1])
```

```python
plt.scatter(W[:,0][z_pred==0], W[:,1][z_pred==0])
plt.scatter(W[:,0][z_pred==1], W[:,1][z_pred==1], color = "c")
plt.scatter(W[:,0][z_pred!=Z], W[:,1][z_pred!=Z], color = "r") # red dots indicate incorrect classification
```
